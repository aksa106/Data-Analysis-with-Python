* Topic 4 [Describe()]

    * desribe() gives - statistical summary of numerical columns
    
    * IT IGNORES TEXT COLUMNS (name,gender,..etc)

    * Pandas only analyzes - integers, floats, numerical data

    * Meaning of all imp terms 

        * count : how many non-empty values exist

        * mean : Average (sum/number of values)

        * std : Standard deviation (how spread out data is)

        * min

        * max

        * Percentiles 

            * 50% : Median (Middle Value)

            * 25% : Lower quarter

            * 75% : Upper quarter

```python

df = pd.read_csv("student.csv")

print(df.describe())
```

* Topic 5 (dtypes)

    * tells data type of every Column

        *  [Imp DataTypes]
        * Object -- string/text
        * int64  -- Integer values
        * float64 -- Decimal numbers




        

    * Why Analysts checks Dtypes?
        * Because wrong dtypes = wrong analysis
        * If salary column becomes object instead of int64
            * averages may fail
            * calculations break
            * sorting becomes weird

```python
df = pd.read_csv("student.csv")

print(df.dtypes)
```

* Q1 What datatype usually represents text/string columns in Pandas?
    * object

* Q2 Which datatype stores decimal values?
    * float64
