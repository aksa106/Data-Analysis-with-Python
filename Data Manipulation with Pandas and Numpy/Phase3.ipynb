#### Topic 1 : Creating New Columns 
    * We can create new column by performing row wise ops.

        * Pandas automatically does operations row by row
        * we do not have to write loops
        * When making a new column by performing an operation
            * you must create a variable and assign the new column values to it 


* CLASSIC BEGINNER MISTAKE 
    * always attach df[Column_name] 
    * Total_Price = creates a python variable
    * df["Total_Price"]  = creates a data frame column

                    

```python
import pandas as pd 

df = pd.read_csv("Phase3.csv")

df["Total_Price"] = df["price"] * df["quantity"]

df["New_Column"] = ["val1","Val2","Val3"]

print(df)
```

```python
df = pd.read_csv("Phase3.csv")

df["profit"] = df["price"] * 0.2

print(df)
```

#### Topic 2 : Renaming Columns 

* df.rename() -- renames columns, indexes
    * Syntax 
        * df.rename(columns = {
            "Old_colName" : "New_colName"
        })

* rename() -- returns a new Data Frame

* CLASSIC MISTAKE 
    * df.rename(columns={"price":"product_price"})
    * print(df)

        * No change will happen and og unchanged data frame will return as output
        * To really make change we must assign it ti df = 

```python
df = pd.read_csv("Phase3.csv")

df = df.rename(columns = {
    "price" : "product_price"
})

print(df)
```

#### Astype()

* Changing Datatypes of Columns 

    * Common Conversions 
        * int - Float
        * Float - int
        * number - str

        #### VERY Important Real-World Use

        * Sometimes CSV columns come wrong.
            * Salary may become object instead of numeric
            * Analysts fix that using astype()

        #### Common Beginner Mistake

        * astype(int) can fail
        * if column contains : text, NaN, invalid values (You must clean data first)


```python


df = pd.read_csv("Phase3(1).csv")

df["age"] = df["age"].astype(float)
## Here We convereted the Data type of column age into float

print(df.dtypes)
```

```python


df = pd.read_csv("Phase3(1).csv")

df["age"] = df["age"].astype(str)

print(df.dtypes)
```

#### Topic 4 -- apply()
* apply a function to column values

    #### Real Usage of apply()
    * apply is used for : 
    * category labeling
    * transformations
    * custom logic
    * scoring systems
    * formatting
    * feature engineering

```python
df = pd.read_csv("Phase3(2).csv")

df["Double_price"] = df["price"].apply(lambda x: x*2)
# made new column    selected col    applies function to every value

# pandas internally applies Ops to every row automatically  
print(df)
```

```python


df = pd.read_csv("Phase3.csv")

df["new_price"] = df["price"].apply(lambda x: x + 100)

print(df["new_price"])
```
