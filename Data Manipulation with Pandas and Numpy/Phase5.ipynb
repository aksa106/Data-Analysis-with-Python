# Merging & Joining
SQL JOIN inside Pandas

### pd.merge  (DataFrame1, DataFrame2, ON = "commonColumn")
* Pandas will merege both the Tables/Data Frames using ON 
    * Only matching rows will come in output (By default inner join)

```python
import pandas as pd 

df1 = pd.DataFrame({
    "id" : [1,2,3],
    "name" : ["A","B","C"]
})


df2 = pd.DataFrame({
    "id" : [1,2,4],
    "salary" : [50000, 60000, 70000]
})


merged = pd.merge(df1, df2, on="id")

print(merged)
```

#### Left Join
* " Keep my main table complete, and attach extra information available "

```python
df1 = pd.DataFrame({
    "id" : [1,2,3],
    "name" : ["A","B","C"]
})


df2 = pd.DataFrame({
    "id" : [1,2],
    "salary" : [50000,60000]
})

merged = pd.merge(
    df1,
    df2,
    on = "id",
    how = 'left'
)

print(f"{df1} \n \n{df2}")

print("\n")

print(merged)
```

### Right join
* Keep ALL rows from right dataframe.

```python
df1 = pd.DataFrame({
    "id" : [1,2,3],
    "name" : ["A","B","C"]
})


df2 = pd.DataFrame({
    "id" : [1,2],
    "salary" : [50000,60000]
})

merged = pd.merge(
    df1,
    df2,
    on = "id",
    how = 'right'
)

print(f"{df1} \n {df2}")

print()

print(merged)
```

#### OUTER JOIN
* It Keeps everything from both of the tables.

```python
df1 = pd.DataFrame({
    "id" : [1,2,3],
    "name" : ["A","B","C"]
})


df2 = pd.DataFrame({
    "id" : [1,2,4],
    "salary" : [50000,60000,70000]
})

merged = pd.merge(
    df1,
    df2,
    on = "id",
    how = 'outer'
)

print(f"{df1} \n {df2}")

print()

print(merged)
```
