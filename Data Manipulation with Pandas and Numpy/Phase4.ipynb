# Aggregation & Grouping 

#### Groupby()
* same as SQL GROUP BY inside Python

* groupby() alone does almost nothing useful
    * You usually combine it with:
    * sum()
    * mean()
    * count()
    * agg()


```python
import pandas as pd 

df = pd.read_csv("Phase4.csv")

grouped = df.groupby("product")["sales"].sum()

## groupby = grouped rows based on product names

## ["sales"] = selected sales column

## .sum() = added sales values , wrt to each categorised product


print(grouped)
```

```python
df = pd.read_csv("Phase4.csv")

grouped = df.groupby("product")["sales"].mean()

print(grouped)
```

```python
df = pd.read_csv("Phase4.csv")

grouped = df.groupby("product")["sales"].mean()

print(grouped)
```

```python
grouped = df.groupby("region")["sales"].sum()

print(grouped)
```

#### agg()
* stands for aggregate 
    * It is used to :
    * apply one or multiple aggregated fucntions on grouped Data.
    * Examples : sum, mean, max, min, count

* Syntax 
    * df.groupby("column_name)["numeric_column].agg([fucntions])

### What Each Metric Means
* sum

    * Total department salary expense.

* mean

    * Average salary.

* max

    * Highest salary in department.

* min

    * Lowest salary.

* count

    * How many employees exist.


    #### THIS Is Real Analysis
* You’re answering business questions:

    * Which department spends most?
    * Which has highest average pay?
    * Which team is larger?

##### That’s why aggregation is core analyst work.




```python
df = pd.read_csv("Phase4(1).csv")

sumamry = df.groupby("department")["salary"].agg(
    ["sum","mean","max","min","count"]
)

print(sumamry)
```

#### Multiple GroupBy 
* Now we group using MORE THAN ONE column.

    #### Why is multiple grouping more detailed than single grouping?
    * --
        * groupby("product") only gives broad summaries like total sales of each product overall. But groupby(["product","region"]) breaks the data further, showing product sales separately inside each region, so the analysis becomes more detailed and specific.

```python
df = pd.read_csv("Phase4.csv")

grouped = df.groupby(["product","region"])["sales"].sum()
#           selected multiple Cols for grouping , selected agg col

# Now grouping is happening on product & region together
print(grouped)
```
