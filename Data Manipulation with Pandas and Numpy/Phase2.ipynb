### Does fillna() automatically change original dataframe?




  * fillna() does not automatically change the original dataframe because, by default, Pandas tries to avoid accidental permanent modifications to your data. When you use fillna(), it creates and returns a new dataframe with the missing values replaced, while the original dataframe stays unchanged unless you explicitly store the result back into a variable like df = df.fillna(0).  

```python
##lets see if missing value still existss or not
df = pd.read_csv("Phase2.csv")

df.fillna(0)

print(df.isnull().sum())

## Missing Values still exists as , Although df.fillna(0) fills the Data Frame but never stores it anywhere, So that is why the we are still getting missing values.
```

```python
df = pd.read_csv("Phase2.csv")

new_df = df.fillna(0)

print(f"Original df   \n{df.isnull().sum()}")

print("\n")

print(f"New df   \n{new_df.isnull().sum()}")
```

* Topic 4 (fillna(mean))
    * Instead of blindly replacing with 0,
     we replace missing values with the column average.

```python
df = pd.read_csv("Phase2.csv")

df["marks"] = df["marks"].fillna(df["marks"].mean())
#1. df["marks"] = selects only marks column
#2. .mean()  = Calculates average marks 
#3. .fillna(df["marks"].mean()) = replaces missing values with mean values 


df["age"] = df["age"].fillna("Unknown")

print(df)
```

#### Why is mean replacement usually better than replacing with 0
* Mean repalcement on missing values is usually better than 0 because 
    * 0 can distort the data and create an unrealistic meaning
    * A missing salary can not be replaced  by zero as it creates an unrealistic meaning as salary can never be zero
    * Using the Mean keeps the data more distributed, balanced and reduces the impact of missing values on analysis

```python
import pandas as pd
df = pd.read_csv("Phase2.csv")

avg = df["marks"].mean()

df["marks"] = df["marks"].fillna(df["marks"].mean())

df["age"] = df["age"].fillna(df["age"].mean())

df["name"] = df["name"].fillna("Unknown")

#df["age"] = df["age"].fillna(0)

print(avg)
print(df)
```
