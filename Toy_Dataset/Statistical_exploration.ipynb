# Measure of Center 

* A measure of center tells us the typical or centeral value of a numerical variavle.
* "If I had to describe this entire column with one representative value, what would it be?"
* For example:

    Typical customer age
    Typical order value

## Types of Measures of Center
* Mean : The arithmetic average. (Average sales amount.)
* Median : The middle value after sorting (Typical customer spending.)
* Mode : The value occurring most frequently.(Most common payment mode, age, or product category.)

```python
import pandas as pd 

import numpy as np

import matplotlib.pyplot as plt 

import seaborn as sns
```

```python
# Load dataset

df = pd.read_csv("Clean_Data.csv")
```

```python
## For Statistical exploration we will just only use 3 columns 
# Age
# Amount
# Orders

num_col = ['Age','Amount','Orders']

```

```python
df[num_col].describe()
```

```python
plt.figure(figsize=(12,4))

for i, col in enumerate(num_col,1):
    plt.subplot(1,3,i)
    sns.boxplot(y=df[col])
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python


num_col = ['Age', 'Amount', 'Orders']

plt.figure(figsize=(15,4))

for i, col in enumerate(num_col, 1):
    plt.subplot(1,3,i)
    plt.hist(df[col], bins=10, edgecolor='black')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(15,4))

for i, col in enumerate(num_col, 1):
    plt.subplot(1,3,i)
    sns.kdeplot(df[col], fill=True)
    plt.title(f'KDE Plot of {col}')

plt.tight_layout()
plt.show()
```

```python
df[num_col].skew()
```

```python
corr = df[num_col].corr()

corr
```

```python
plt.figure(figsize=(6,5))

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title('Correlation Heatmap')
plt.show()
```
