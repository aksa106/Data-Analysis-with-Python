# Data Cleaning

### Remove Unecessary Columns
* There are 2 columns which are filled with nulls and contribute nothing to our analysis
    * Status : Contains nulls only
    * notes : Contains nulls only

* Removing  both columns as they provide no analytical value.

```python
df[['Status','Notes']].info()
```

```python
# dropping cols : status and notes

df.drop(columns=['Status','Notes'], inplace=True)
```

```python
# Revalidating

df.columns
```

### Identifier Columns Validation

* Identifier Columns : Customer_Id


```python
# * As Customer_ID data type is int, because it represents labels not numeric values, and does not contributes to any numerical analysis

df['Customer_ID'] = df['Customer_ID'].astype(str)
```

```python
df['Customer_ID'].dtype
```

```python
## Nulls Check

df['Customer_ID'].isnull().sum()

# No nulls in Customer_ID column
```

```python
## Duplicate check

df['Customer_ID'].duplicated().sum()

# Duplicates :  2

df[df['Customer_ID'].duplicated(keep=False)]
```

```python
## Dropping Duplicate Columns

df.drop_duplicates(subset=['Customer_ID'], inplace=True)
```

```python
## Revalidation

df['Customer_ID'].duplicated().sum()
```

```python

# Handling nulls

categorical_cols = [
    'Customer_Name',
    'Gender',
    'State',
    'Product',
    'Category',
    'Payment_Mode'
]

df[categorical_cols].isnull().sum()
```

```python
df[df[['Gender','State']].isnull().any(axis=1)]
```

#### Category Validation — Consistency & Standardization

* Til now, i have removed all of the nulls and duplicates from categorical cols , let us investigate data consistency and categories.

```python
categorical_cols = [
    'Customer_Name',
    'Gender',
    'State',
    'Product',
    'Category',
    'Payment_Mode'
]

for col in categorical_cols:
    print(f'\n{col}')
    print(df[col].unique())

## As we can see there are data inconsistency in  2 columns
# Gender contains inconsistent labels.
# State contains spelling inconsistency.
```

```python
## CLeaning data inconsistency

df['Gender'] = df['Gender'].replace({
    'M' : 'Male',
    'F' : 'Female'
})
```

```python
df['State'] = df['State'].replace({
    'Maharastra':'Maharashtra'
})
```

```python
## Revalidation 

categorical_cols = [
    'Gender',
    'State'   
]

for col in categorical_cols:
    print(f'\n{col}')
    print(df[col].unique())
```

```python
numerical_cols = ['Amount','Age','Orders']

df[numerical_cols].isnull().sum()
```

```python
## Handling Nulls 

df.dropna(subset=['Amount'], inplace=True)
```

```python
# Revalidate 

df['Amount'].isnull().sum()
```

```python
## Handling duplicates
numerical_cols = ['Amount','Age','Orders']

df[numerical_cols].duplicated().sum()

## There are no duplicates
```

```python
## Now validating Numerical Data for negative values 

df[df['Amount']<0]
```

```python
## let us drop this col as it does not makes sense someone buyins a short for -500 rupees

df.drop(index=38, inplace=True)
```

```python
## Revalidation

df[df['Amount'] < 0]
```

```python
## Outlier investigation , as we saw in statistical summary we had, an absurdly large amount

df[df['Amount'] == df['Amount'].max()]

## As we can see it does not makes any sense someone buying 3 shirts for 999999, it is clearly a data entry mistake 

# the maximum transaction amount is unrealistic and is a data entry error
```

```python
## Let us remove this column

df.drop(index=48, inplace=True)
```

```python
## Revalidation

df[df['Amount'] == df['Amount'].max()]
```
