# Matplotlib Foundations 


* WHAT IS MATPLOTLIB ?
    * A Python library used to turn Data - Graphs.
    * Data -> Visualization 
        * Examples : 
            * Line chart
            * Bar Chart
            * Histogram
            * Scatter plot

    * Matplotlib is a visualization library , it is like a plotting toolbox
        * It has
        * many modules
        * many functions
        * pyplot -> plotting toolbox




### pyplot 
* pyplot is a plotting toolbox inside matlpotlib, Most graphs are created using it.

* plt.plot(x,y)
---------------------------------------------------------------------


### Labels 
* Names of the axes

* plt.xlabel("Label_name")
---------------------------------------------------------------------

### Title 

* plt.title(" Title_name ")

---------------------------------------------------------------------
* plt.plot()    # create graph
* plt.xlabel()  # horizontal axis name
* plt.ylabel()  # vertical axis name
* plt.title()   # graph title

---------------------------------------------------------------------

### GRID 
* Without a grid , estimating values is harder
* plt.grid(True)

--------------------------------------------------------------------

### COLORS

* By default matplotlib chooses a color, but we can also choose our own

---------------------------------------------------------------------

### MARKERS

* Rn we only see the plotted line but sometimes we need to see the, actual data points too.

---------------------------------------------------------------------

## Line Style

* Solid "-"
* Dashed line "--", 
* Dotted line ":"
* Dash-dot line "-."

---------------------------------------------------------------------

## Line width

* line width = 4

---------------------------------------------------------------------

## Marker size 
* markersize = 12


```python
import matplotlib.pyplot as plt 

# Let's create some data to plot 

x = [1,2,3,4,5,6]

y = [1,4,9,16,25,36]   

plt.plot(x,y) 
#"Plot the x values on the horizontal axis and the y values on the vertical axis, then connect the points with lines."

#---------------------------------------------------------------------------

## LABELS

plt.xlabel("Age")
plt.ylabel("Salary")

## TITLE -----------------------------------------------------------------------

plt.title("Age Vs Salary")


## GRID---------------------------------------------------------------------------

plt.grid(True)


## COLOR  & MARKER & LINE STYLE & LINE WIDTH & MARKER SIZE ---------------------------------------------------------

plt.plot(
    x,y, 
    color = 'red', 
    marker = "o",
    linestyle = "-",
    linewidth = 3,
    markersize = 9
    )


plt.show()

```
