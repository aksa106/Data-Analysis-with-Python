# TOPIC 2 : FIRST REAL HISTOGRAM 

* Let us create a sales distribution histogram, 
    We have Category, Date, value, product, sales , region

    we are creating a sales distribuiton to see the sales values into ranges and counts them

```python
import pandas as pd , matplotlib.pyplot as plt

df = pd.read_csv("data.csv")

df["Sales"].hist()

plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")

plt.show()
```

## TOPIC 3 : FIRST REAL BAR CHART 

#### Which product generated more total sales?


```python
import pandas as pd, matplotlib.pyplot as plt 

df = pd.read_csv("data.csv")

product_sales = df.groupby("Product")["Sales"].sum()

product_sales.plot(kind = "bar")

plt.title("Total Sales per product")

plt.xlabel("Product")

plt.ylabel("Total Sales")

plt.show()
```

#### MEAN SALES

```python
import pandas as pd , matplotlib.pyplot as plt 

df = pd.read_csv("data.csv")

Mean_sales = df.groupby("Product")["Sales"].mean()

Mean_sales.plot(kind="bar")

plt.title("Mean Sales")

plt.xlabel("Products")

plt.ylabel("Mean")

plt.show()
```

### NUMBER OF PRODUCTS

```python
import pandas as pd , matplotlib.pyplot as plt 

df = pd.read_csv("data.csv")

NumberOfProducts = df.groupby("Product")["Value"].count()

NumberOfProducts.plot(kind="bar")

plt.title("Number of Products")

plt.show()
```
