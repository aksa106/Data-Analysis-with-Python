# TOPIC 2 : FIRST REAL HISTOGRAM 

* Let us create a sales distribution histogram, 
    We have Category, Date, value, product, sales , region

    we are creating a sales distribuiton to see the sales values into ranges and counts them

```python
import pandas as pd , matplotlib.pyplot as plt

df = pd.read_csv("data.csv")

df["Sales"].hist()

plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")

plt.show()
```

## TOPIC 3 : FIRST REAL BAR CHART 

#### Which product generated more total sales?


```python
import pandas as pd, matplotlib.pyplot as plt 

df = pd.read_csv("data.csv")

product_sales = df.groupby("Product")["Sales"].sum()

product_sales.plot(kind = "bar")

plt.title("Total Sales per product")

plt.xlabel("Product")

plt.ylabel("Total Sales")

plt.show()
```

#### MEAN SALES

```python
import pandas as pd , matplotlib.pyplot as plt 

df = pd.read_csv("data.csv")

Mean_sales = df.groupby("Product")["Sales"].mean()

Mean_sales.plot(kind="bar")

plt.title("Mean Sales")

plt.xlabel("Products")

plt.ylabel("Mean")

plt.show()
```

### NUMBER OF PRODUCTS

```python
import pandas as pd , matplotlib.pyplot as plt 

df = pd.read_csv("data.csv")

NumberOfProducts = df.groupby("Product")["Value"].count()

NumberOfProducts.plot(kind="bar")

plt.title("Number of Products")

plt.show()
```

## Real Scatter Plot 

* Goal : See of two numerical columns are related 

#### From data.csv -- Value and Sales are related , When value increases does Sales also increases.




```python
import pandas as pd , matplotlib.pyplot as plt 

df = pd.read_csv("data.csv")

plt.scatter(df["Value"], df["Sales"])

plt.title("Value vs Sales")

plt.xlabel("Value")

plt.ylabel("Sales")

plt.grid(True)


plt.show()


##  Dots Rise upwards , So that is a Positive relationship

## Dots everywhere -- No clear relationship



## in our graph dots are scattered randomly, so there is neither a strong upward trend, nor a strong downward trend.

## Analyst Conclusion -- No obvious relationship between value and sales
```

## Real Pie Chart 

* Which Region appears most often in the dataset?

    GOAL : Find which region appears most often 
    



```python
import pandas as pd , matplotlib.pyplot as plt 

## Step 1 : Count the regions 

df = pd.read_csv("data.csv")

region_count = df.groupby("Region").count() ## Groups by region and counts records

##Plotting now

plt.pie(
    region_count["Sales"],  ## Selects the sales column 
    labels = region_count.index, ## region_count.index = ["East",West,North]

    # use east west north as the names of the pie slices

    autopct="%1.1f%%"
)

plt.title("Region Ditribution")

plt.show()
```

# REAL TIME - SERIES LINE PLOT 

* How do Sales changes over time 

    * Step 1 : check your date column

```python
import pandas as pd , matplotlib.pyplot as plt 

df = pd.read_csv("data.csv")


## Step 1 : Check your Date Column 

#print(df["Date"])


## Step 2 : Create total sales per day

sales_trend = df.groupby("Date")["Sales"].sum()

#print(sales_trend)



## now plot it 

plt.plot(
    sales_trend.index,
    sales_trend.values
)

plt.title("Sales Trend Over Time")

plt.xlabel("Date")

plt.ylabel("Sales")

plt.grid(True)

plt.show()

## Sales jump around day to day
```
