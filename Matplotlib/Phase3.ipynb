# TOPIC 1 : BAR CHART


## Bar Plot 
* plt.bar(values) 
    * creates a bar chart based upon given values

        * LABEL AND TITLE ON BAR CHARTS 
            * plt.xlabel
            * plt.ylabel

        * BAR COLORS 
            * plt.bar(color = "color")

        * INDIVIDUAL BAR COLORS 
            * plt.bar(color = ["green","red","blue"])


```python
import matplotlib.pyplot as plt 

products = ["Laptop", "Phone", "tablet"]

sales = [100,150,80]

plt.bar(products,sales,color = ["blue","red"])


## GIVING LABELS 

plt.xlabel("Products")
plt.ylabel("Sales")

## GIVING title
plt.title("Product Sales")


## Giving Grid
plt.grid(True)


plt.show()
```

#### MINI CHALLENEGE
* Create a bar chart for:
-------------------------------------------------------------------------------
* students = ["Arun", "Rahul", "Priya"]

* marks = [85, 72, 91]

-------------------------------------------------------------------------------

Requirements:

    Bar chart
    Title = "Student Marks"
    X label = "Students"
    Y label = "Marks"
    Color = orange
    Grid enabled

# TOPIC 2 : HISTOGRAM

* Bar Chart asks:

    Which category is bigger?

* Histogram asks:

    How is my data distributed?

    * Histogram groups values into ranges and then counts 

        * Suppose you have : marks = [45, 50, 55, 60, 62, 65, 70, 72, 75, 80, 85, 90]
        * Q : how are these values of marks spread out? 

        * histograms groups values into ranges and then counts.

```python
import matplotlib.pyplot as plt 

marks = [45, 50, 55, 60, 62, 65, 70, 72, 75, 80, 85, 90]

plt.hist(marks, bins = 10) # bins = no. of ranges

plt.show 
```

#### TINY HISTOGRAM CHALLENEGE

```python
import matplotlib.pyplot as plt

scores = [45, 52, 55, 60, 62, 65, 68, 70, 72, 75, 80, 85, 90]

plt.hist(scores, bins=5)

plt.xlabel("Scores")
plt.ylabel("Frequency")

plt.grid(True)

plt.title("Exam Scores Distribution")

plt.show()
```

#  TOPIC 3 : SCATTER PLOT

### What problem does it solves?

* Suppose you have : 
    hours = [1,2,3,4,5]

    marks = [40,50,60,75,90]

* Question : Do students who study more score higher?
    * Scatter plots are ususally used to study relationships.
    
* Use scatter plot when u see:
    * Two numerical variables are related.



```python
import matplotlib.pyplot as plt 

hours = [1,2,3,4,5]

marks = [40,50,60,75,90]

plt.scatter(hours,marks)

plt.title("Study hours vs Marks")

plt.xlabel("Hours Studied")

plt.ylabel("Marks")

plt.grid(True)

plt.show()
```

# TOPIC 4 : PIE CHART

### what problem does it solve? 

* Show parts of a whole
    * shows how the whole is divided

    plt.pie(values, labels=labels)

```python
import matplotlib.pyplot as plt 

products = ["Laptop", "Phone", "Tablet"]

sales_share = [90, 35, 25]

plt.pie(sales_share, labels = products)

plt.show()
```

```python
import matplotlib.pyplot as plt

products = ["Laptop", "Phone", "Tablet"]

sales_share = [40, 35, 25]

plt.pie(
    sales_share,
    labels = products,
    autopct = "%1.1f%%" # shows percentage on pie chart
)

plt.title("Sales Share")


plt.show()
```

## Tiny Challenge

    Create a pie chart for:
    subjects = ["Python", "SQL", "Excel"]

    time_spent = [50, 30, 20]

    Requirements:

    Pie chart
    Labels
    Percentages displayed
    Title = "Study Time Distribution"

```python
import matplotlib.pyplot as plt 

subject = ["Python", "SQL", "Excel"]

time_spent = [50,30,20]

plt.pie(time_spent,labels = subject, autopct= "%1.1f%%")

plt.title("Study Time distribution")

plt.show()
```
