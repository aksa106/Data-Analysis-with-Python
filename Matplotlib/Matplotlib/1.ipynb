```python
import matplotlib.pyplot as plt
import seaborn as sns
```

* plt.figure()  : creates an empty fogure/ canvas where graph will be drawn
* figsize=(x,y) : controls how much large the figure/canvas will be
    * (6,4) : defualt small plots
    * (8,5) : Most EDA charts
    * (10,6) : Long/labels

* Figure : entire window/canvas
* Axes : The actual area where the graph is drawn
* plt.bar(x,y) : creates a bar chart to compare categories

* plt.title()
* plt.x,ylabels()

* plt.x,yticks(fontsize, rotation= 45) : use rotation when category labels overlap to improve readability.

* legend : a legend identifies the meaning of different colors lines, markers in a  visualiZation.
    * plt.legend(loc='upper left)
    * 'upper right'
    * lower right
    * best : chooses the best position automatically

* grid : Grids improve readability by making it easier to estimate values from the chart. They don't change the data.
    * plt.grid(axis="y")
    * plt.grid(axis="x")
    * plt.grid(linestyle="--")
    * for bar charts -> axis="y" (helps compare heights)
    * for line charts -> For line charts → Full grid is often useful.

* plt.savefig() : Saves the chart as image 
* Save them as (.png, .jpg, .pdf, .svg)

# Basic Matplotlib Plots

* plt.plot() -- Line Chart 
    * shows trends over an ordered sequence , usually time
    * Example : monthly sales, daily revenue, website visitors
    * plt.plot(x,y)
    * Use when: X-axis has an order (time, sequence, progression).

* plt.scatter() -- Scatter Plot
    * Shows the relationship between two numeric variables.
    * Example: (Age vs Salary, Height vs Weight , Spend vs Sales)


| `plt.plot()`         | `plt.scatter()`       |
| -------------------- | --------------------- |
| Shows trends         | Shows relationships   |
| Points are connected | Points are separate   |
| Time-series data     | Two numeric variables |


* plt.hist() -- Histogram
    * shows the distribution of a numeric variable
    * Age, salary, product prices
    * frequency distrbution, how many times one value occurs

* plt.pie() -- Pie chart
    * shows parts of a whole as percentages
    * Payment Mode, Market Share, Browser Usage

| `plt.hist()`                 | `plt.pie()`              |
| ---------------------------- | ------------------------ |
| Distribution of numeric data | Part-to-whole comparison |
| Numeric values               | Categorical percentages  |
| Shows frequency              | Shows proportions        |



* plt.barh() -- Horizontal bar chart
    * A horizontal version of bar chart
    * use it when category names are long.

* plt.boxplot() -- shows the spread of data and helps detects outliers 





| `plt.barh()`          | `plt.boxplot()`              |
| --------------------- | ---------------------------- |
| Compare categories    | Analyze numeric distribution |
| Horizontal bars       | Detect spread & outliers     |
| Categorical + Numeric | One numeric variable         |



```python
import matplotlib.pyplot as plt


age = [20,21,22]

height = [160,160,165]

plt.scatter(age,height)

plt.title('Age Vs Height')

plt.show()
```

```python
import matplotlib.pyplot as plt

age = [20, 21, 22, 22, 23, 25, 26, 28, 30, 35, 40]

plt.hist(age)
plt.title("Age Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt

payment = [40, 35, 25]
labels = ["UPI", "Card", "Cash"]

plt.pie(payment, labels=labels)
plt.title("Payment Mode Distribution")
plt.show()
```

```python
import matplotlib.pyplot as plt

states = ["Maharashtra", "Uttar Pradesh", "Karnataka"]
sales = [120, 95, 80]

plt.barh(states, sales)
plt.title("Sales by State")
plt.show()
```

```python
import matplotlib.pyplot as plt

amount = [200, 220, 250, 270, 300, 320, 350, 1200]

plt.boxplot(amount)
plt.title("Order Amount Distribution")
plt.show()
```
