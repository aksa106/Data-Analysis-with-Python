```python
import matplotlib.pyplot as plt



products = ['laptop','phone','tablet']

sales = [120, 80, 45]

plt.figure(figsize=(6,4))
plt.bar(products, sales)

plt.title(
    'Sales Comparison',
    fontweight = 'bold',
    fontsize = 16,
    loc='center'
)

plt.xlabel(
    'Products',
    fontsize=14

    )


plt.ylabel(
    'Sales',
    fontsize= 14
    )

plt.grid(axis='y')
plt.show()
```

```python
import matplotlib.pyplot as plt 

dept = ['HR', 'Finance', 'Sales', 'IT', 'Marketing']

Emp = [12, 18, 35, 28, 20]


plt.figure(figsize=(6,4))

plt.barh(dept, Emp)

plt.title(
    'Number of Employees in each Dept',
    fontweight='bold',
    fontsize = 14,
    loc='center'
    )

plt.ylabel('Departments', fontsize=13)
plt.xlabel('No. of Employees', fontsize=13)

plt.grid(axis='x')

plt.show()


```

```python
import matplotlib.pyplot as plt


months = ['Jan', 'Feb', 'March', 'Apr', 'May','Jun']

revenue = [25, 32, 28, 40, 45, 50]


plt.figure(figsize=(6,4))

plt.plot(months, revenue)

plt.title(
    'Revenue Trend',
    fontweight='bold',
    fontsize=16,
    loc='center'
    
    )

plt.xlabel('Months', fontsize=13)

plt.ylabel('Revenue', fontsize=13)

plt.grid(axis='y')


plt.show()
```

```python
age = [22, 25, 28, 31, 35, 40]

salary = [25000, 32000, 41000, 52000, 65000, 78000]


plt.figure(figsize=(6,4))

plt.scatter(age, salary)

plt.title(
    'Age Vs Salary',
    fontweight='bold',
    fontsize=16,
    loc='center'
    )

plt.xlabel('Age',fontsize=14)
plt.ylabel('Salary',fontsize=14)

plt.grid(True)

plt.show()
```

```python
marks = [45, 52, 60, 61, 65, 67, 70, 72, 74, 75,
         78, 80, 82, 85, 88, 90, 91, 93, 95, 98] ## n = 20



plt.figure(figsize=(6,4))

plt.hist(marks)
plt.title(
    'Student Marks',
    fontweight='bold',
    fontsize=16,
    loc='center'
    
    )

plt.xlabel('marks', fontsize=14 )
plt.ylabel('Frequency', fontsize=14)

plt.grid()


plt.show()





```

```python

import matplotlib.pyplot as plt

categories = ['Electronics', 'Clothing', 'Grocery', 'Furniture']
sales = [40, 25, 20, 15]

plt.figure(figsize=(6, 6))

plt.pie(sales, labels=categories)

plt.title(
    'Sales Distribution by Category',
    fontsize=16,
    fontweight='bold',
    loc='center'
)

plt.show()



```
