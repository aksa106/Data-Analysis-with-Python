# Multi Plots & Layouts

## Figure()
* plt.plot creates a graph
    * but if you want 
        * multiple graphs
        * bigger graphs
        * smaller graphs
        * better layout 
* We use figure() -- we first create a canvas and that canvas is called figure()

* figure
* Graph 1
* Graph 2 
* Graph 3  
    * Figure is the container
    * Figure = Empty room
    * Axes = Actual graph inside room


-------------------------------------------------------------------------------

## Figsize
* Sometimes graphs are too small 
* You can control thier size
    * width = 8
    * height = 4



    * FIGURE ()  = CREATES THE PAPER
    * FIGSIZE()  = CHOOSES HOW BIG THE PAPER IS

    * FOR LEARNING THE DEFAULT IS (8,4)



-------------------------------------------------------------------------------

### Subplot 

* Problem 
* Suppose you have 
    * Sales Graph
    * Profit graph 

        * Do you want : 
            * Window 1 -- Sales
            * Window 2 -- Profit

        * one window 
            * --- Sales
            * --- Profit

* Subplot Divides one figure into sections 


* figure()  → create paper
* figsize() → choose paper size
* plot()    → draw on paper

-------------------------------------------------------------------------------


```python
### FIGURE

import matplotlib.pyplot as plt 

x = [1,2,3]

y = [1,4,9]

plt.plot(x,y)

plt.title("Small")

plt.figure(figsize=(4,2)) ## This creates an empty canvas (Blank sheet of paper)

plt.show()


```

```python
### FIGURE

import matplotlib.pyplot as plt 

x = [1,2,3]

y = [1,4,9]

plt.plot(x,y)

plt.title("Large")

plt.figure(figsize=(12,6)) ## This creates an empty canvas (Blank sheet of paper)

plt.show()
```

### subplot

### GOAL 

* One graph  

    * to

* Multiple graphs on paper

```python
import matplotlib.pyplot as plt 

# (1,2,1) - First Number = Rows = 1 row
#  Second Number = Columns = 2 columns
# Third Number = Current Slot (Use slot #1)


## 1 row , 2 boxes , use box 1

plt.subplot(1,2,1)  

plt.show()
```

```python
plt.subplot(1,2,2)

## 1 row 2 boxes use box 2

plt.show()
```

```python
import matplotlib.pyplot as plt 

plt.subplot(1,2,1)

plt.plot([1,2,3],[1,4,9])

plt.title("Graph1")

########################################################################

plt.subplot(1,2,2)

plt.plot([1,2,3],[9,4,1])

plt.title("Graph2")
```
