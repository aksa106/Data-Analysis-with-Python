# Control Flow and Program Structure


### Multiple Parameters

```python
## A function can take multiple parameters

## Ordered Parameters 

## Positional Arguments 

def subtract(a,b):
    return a-b 

subtract(3,6) 

## Arguments will take exact same positions as parameters 
## In this, we got to order both our parameters and arguments accordingly and meaningfully
```

```python
## Keyword arguments : Order does not matter

##Let us take an example 

def info(city, name):
    return f"{name} lives in {city}"

info(city="Delhi", name="Xyz")

info(city="Xyz", name="Delhi")

info("xyz","Delhi")


## As we can see , if we dont specify what parameters takes what argument it automatically gives the positional arguments based on position of paramter

## if we specify using keywords, the order wont matter as it will take arguments based on keywords only

```

```python
## Deault parameters 

#Lets take an example 

def greet(name="xyz", country="India"):
    return f"{name} lives in {country}"

greet(name="abc", country="IN")

greet()


## Defualt paramters are those paramters which are pre defined in fucntion defination statement , we specify values for respective parameters

## if any argument is not being passed in callback fucntion, the defualt parameters gives defualt arguments which were pre defined.



```

```python
## PARAMETERS WITH DEFAULT VALUE SHOULD COME AFTER NON DEFAULT PARAMETERS 

def greet(name, country="in"):  ## default parameter came afterwards
    return f"{name} lives in {country}"

greet("abc")
```

```python
## defualt paramter came first

def greet(country="in", name):  
    return f"{name} lives in {country}"

greet("abc")

## Syntax error : non default parameter came after a default one
```

```python
## Returning Multiple Values

def calculate(a,b):
    return a+b, a-b, a*b, a/b, a**b

sum, diff, prod, div, exp = calculate(10,20)

print(f" {sum}\n {diff}\n {prod}\n {div}\n {exp}")

## A fucntion can return more than one value using a single return statement , 
## Python automatically packs the returned value into a tuple, and you can unpack them into separate variabels
```

## Variable Scope (Local vs Global)

Scope determines where a variable can be accessed.

* Global Scope : A global variable can be accessed from anywhere in the program , and it must be created outside all of the fucntions.

* Local Scope : It is created inside of an fucntion and it exists only there only.

```python
country = "India"

def greet():
    name = "xyz"
    print(name)
    print(country)

greet()
print(country)

# name is accessible only inside greet().
# country is accessible both inside and outside the function.

# Important: A local variable cannot be accessed outside its function.
```
