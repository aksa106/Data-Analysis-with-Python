#### Parameters
* A parameter is a variable inside the fucntion defination
    * def func(parameter1, parameter2):
    *   code

* An argument is the actual value you pass when calling the fucntion.
    * def func(parameter1, parameter2):
    *  code
    
    func(argument1, argument2) #These values gets assigned to the parameters


### Parameter : variable in the fucntion
### Argument : actual value in func call

```python
def greet(name):  
    print(f"Hello {name}")

greet("Alice")
greet("bob")
```

```python
def sum(a,b): #sum= function name, a,b = parameters/variables
    sum = a + b
    return sum

sum(2,3)#sum() -- calling the func  , 2,3 = actual arguments
```

### Return Values
* Return sends a value back to the place where function ws called.
* Returned value can be stored in a variable unlike print that just displays output and printed value cannot be stored in a variable.
* Returned value can be used in calculations, whereas printed value only get displayed on console and it cannot be used in calculations.
* returned value can be stored and passed to anothe function, unlinke printed values.
* returned values can be sent in an API response.

### That is why applications rely heavily on returned values.

## Print -- when you want user to see something.
## Return -- when you want your program to use that value.


```python
## Suppose you want to multiply the sum by 2 

def add(a,b):
    print(a+b)

result = add(2,3)*2  
print(result)
## It will raise an error as add(2,3) returns none 
## unsupported operand type(s) for *: 'NoneType' and 'int'
```

```python
def add(a,b):
    return(a+b)

result = add(1,2)*2
print(result)

##returned value = 3 got stored in a variable and used in further calculation unlinke print
```

```python
def add(a, b):
    print(a + b) ## will just display a value

x = add(2, 3)
print(x)  ## will return none
```

```python
def is_even(a):
    if a % 2 == 0:
        return f"{a} is even"
    else:
        return f"{a} is odd"

number = is_even(int(input("Enter your number : ")))
print(number)
```

```python
#Better version

def is_even(number): ## func name = is_even
    return number % 2 == 0  ## code to execute , returned value if val is even


number = int(input("enter your number  : ")) ## dynamic parameter(number)

## logic on fucntion is_even
## Caller is deciding what to do with the result by applying condiitons

##  func(parameter)
if is_even(number):  ## if is_even gets executed and val is even 
    print(f"{number} is even")

else :  ## if is_evne fails and val is odd
    print(f"{number} is odd")


## is_even has one responsibility -- determine whether number is even or odd
## if number is even , stored value gets retuened to the caller
```
