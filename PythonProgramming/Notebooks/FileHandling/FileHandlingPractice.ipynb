### Practice 1 — Daily Tasks Viewer

Scenario

You have a file called tasks.txt.

Contents:

Wake up
Study Python
Go for Walk
Read Book


* Task 1 : wrap eveything inside a fucntion
* Task 2 : Print the heading
* Task 3 : Display numbering
* Task 4 : Remove the extra blank lines

```python
def daily_tasks():

    with open("tasks.txt","r")as file:
        lines = file.readlines()

    print("Todays Task\n")

    i = 1
    for line in lines:
        print(f"{i}. {line.strip()}") ## .strip() : removed extra blank spaces

        i += 1

daily_tasks()





```

#### Daily Journal
    Scenario

    You have a file called:

    journal.txt

    Current contents:

    Learned Strings
    Learned Lists

    Today you completed File Handling, and you want to save it


    Your Task

    Create a function:

    Requirements
    Part 1 — Add Today's Learning

    Append the following line to the file:

    Display the Updated Journal

    Print:

    Today's Learning Journal

    Then read the file and display:

    1. Learned Strings
    2. Learned Lists
    3. Learned File Handling


```python
def daily_journal():

    with open("journal.txt", "a") as file:
        file.write("\nLearned File Handling")

    with open("journal.txt", "r") as file:
        lines = file.readlines()

    print("Today's Learning Journal\n")

    i = 1
    for line in lines:
        print(f"{i}. {line.strip()}")
        i += 1


daily_journal()
```

```python

def student_report():

    marks = [85,92,78,88,95]

    ## write the file
    with open("marks.txt","w") as file:
        for marks in marks :
            file.write(f"{marks}\n")


    ## read the file

    with open("marks.txt","r") as file:
        lines = file.readlines()


    # Display the report 

    print("REPORT\n")

    i=1
    for line in lines:
        print(f"{i}. {line.strip()}")

        i += 1 

student_report()

  
```
