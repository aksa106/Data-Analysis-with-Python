### File Modes

* Read Mode ("r") : Opens an existing file to read its contents.
    * open("notes.txt,"r")

* Write Mode ("w") : Creates a new file and overwrites its contents.
    * open("notes.txt","w")

* Append Mode ("a") : Adds new content to end of te exisiting file.
    * open("notes.txt","a")

* Create Mode ("x") : Creates a new file only if it doesn't already exists.
    * open("notes.txt","x")

## Opening and closing Files

```python

## open()

#opens a file object and returns a file object that python uses to work with that file.

file = open("notes.txt", "r")

print(file)




# <_io.TextIOWrapper name='notes.txt' mode='r' encoding='utf-8'>

# open() does not read the file. It only opens a connection.
```

```python
## File Object 

# A file object is Python's connection (or handle ) to an opened file.buffer


file = open("notes.txt","r")

#here file is not the file contents, It is an object that lets Python : Read, write , move through and close the file 
# File = Notebook

# File Object = Your hand holding the notebook
```

```python
## close()

# Closes the file and releases the connection.

file = open("notes.txt","r")

file.close()

# Why close()?
# Frees system resources.
# Saves pending changes (when writing).
# Prevents file corruption or locking issue

```

```python
## with open 

# The safest and most recommended way to open files.

# Python automatically closes the file after the block finishes.

with open("notes.txt","r") as file:
    print(file.read())

## no need to write file.close()


## with automatically opens and closes files, even if exception occurs


```
