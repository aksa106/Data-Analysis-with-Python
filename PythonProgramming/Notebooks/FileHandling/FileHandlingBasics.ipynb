### File Modes

* Read Mode ("r") : Opens an existing file to read its contents.
    * open("notes.txt,"r")

* Write Mode ("w") : Creates a new file and overwrites its contents.
    * open("notes.txt","w")

* Append Mode ("a") : Adds new content to end of te exisiting file.
    * open("notes.txt","a")

* Create Mode ("x") : Creates a new file only if it doesn't already exists.
    * open("notes.txt","x")

## Opening and closing Files

```python

## open()

#opens a file object and returns a file object that python uses to work with that file.

file = open("notes.txt", "r")

print(file)




# <_io.TextIOWrapper name='notes.txt' mode='r' encoding='utf-8'>

# open() does not read the file. It only opens a connection.
```

```python
## File Object 

# A file object is Python's connection (or handle ) to an opened file.buffer


file = open("notes.txt","r")

#here file is not the file contents, It is an object that lets Python : Read, write , move through and close the file 
# File = Notebook

# File Object = Your hand holding the notebook
```

```python
## close()

# Closes the file and releases the connection.

file = open("notes.txt","r")

file.close()

# Why close()?
# Frees system resources.
# Saves pending changes (when writing).
# Prevents file corruption or locking issue

```

```python
## with open 

# The safest and most recommended way to open files.

# Python automatically closes the file after the block finishes.

with open("notes.txt","r") as file:
    print(file.read())

## no need to write file.close()

##If you forget close() or an error occurs before it, the file may remain open
## with automatically opens and closes files, even if exception occurs


```

### Reading Files

Learn the different ways to read data from a file.

```python
## read() 

## READS THE ENTIRE CONTENTS OF A FILE AS A SINGLE STRING.

with open("notes.txt","r") as file:
    data = file.read()

print(data) ## Best when File is small
```

```python
## read(Size)

# READS ONLY THE SPECIFIC NUMBER OF CHARACHTERS FROM A FILE.

with open("notes.txt","r") as file:
    print(file.read(6))

#It reads characters, not words or lines.
#because it also counts spaces and newline characters.
```

```python
## readline()

# read one line from the file 

with open("notes.txt","r") as file:
    print(file.readline())
    print(file.readline())
    print(file.readline())


## Each call moves to the next line
```

```python
## readlines()

## Reads all lines and returns them as a list 


with open("notes.txt","r") as file:
    lines = file.readlines()

for line in lines:
    print(line.strip())

print(type(lines))


## Each line becoms one element of the list
```

```python
## File pointer  # intro

# A file pointer keeps track of your current position inside the file

with open("notes.txt", "r") as file:
    print(file.read(7))
    print(file.read())

```

# Writing Files

```python
# write string to a File 

with open("notes.txt","w") as file:
    file.write("Javascript")

    print(file)
```

```python
## write mode


## Opens a file for writing 
# creates a new file if file does not exsits
# clears all previous content in an exisiting file and overwrites it with new content, if new file then writes the content


with open("notes.txt", "w") as file:
    file.write("Power BI")

## now notes.txt will become  Power BI

```

```python
with open("notes1.txt","w") as file:
    data = file.write("this is a new file")

with open("notes1.txt","r") as file:
    data = file.read()

print(data)
```

```python
# Append mode : "a" Mode

# Adds new content at the end of the file without deleting existing data.

with open("notes1.txt","a") as file:
    file.write("\nFlask")


with open("notes1.txt","r") as file:
    data = file.read()

print(data)
```

```python
## writelines

# write multiple strings from a list to a file

courses = [
    "Python\n",
    "SQL\n",
    "Flask\n"
]

with open("notes.txt", "w") as file:
    file.writelines(courses)
    
```
