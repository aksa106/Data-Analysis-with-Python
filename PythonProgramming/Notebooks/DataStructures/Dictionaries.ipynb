### Dictionaries 

* A dictionary stores key-value pairs.
* Dictionary are mutable.
* Dictionary are ordered.
* Keys must be unique , and keys must be immutbale (str, tuple) hashable types.
* Values can be any data type.
* len() counts key-value pairs not just values.

### List Vs Dict 

* List : access by index, [10,20,30] , orderded , mutable 
* Dict : access by key , {"name":"A"} , orderded , mutable

    The biggest mindset shift is this:

    Lists answer: "What is the item at position 2?"
    Dictionaries answer: "What is the value for the key 'name'?"

```python
D = {
    "name":"A",
    "Age" : 22,
    "Gender":"Female"
}

print(D)

print(f"Length of dict (no. of keys) : {len(D)}")
```

### Modify

```python
D = {
    "name":"A",
    "age":22,
    "Gender":"female",
    "place":"Delhi"
}

## Add Data : add a new key-value pair.

D["Country"] = "india"

print(D)




## Update value : Assign a new value to existing key.

D["Place"] = "New Delhi"
print(D)

## update () : update multiple key value pairs or update exisiting ones,

D.update ({
    "Age2" : 21,
    "Place2" : "Delhi"

})
print(D)

## pop() : removes a key value pair and returns the removed value


D.pop("Age2")
D.pop("Place2")
print(D)

## popitem() : Removes and returns the last inserted key-value pair.

D.popitem()
print(D)

## del : Deletes a key-value pair or the entire dictionary.
del D["name"]
print(D)
del D 

## clear() : Removes all key value pairs
D.clear()
print(D)

# ## copy() : create a shallow  copy of dict
newD = D.copy()
print(newD)
```

## Iteration

```python
D = {
    "name": "A",
    "age": 22,
    "gender": "Female",
    "City": "Delhi",
    "state": "Delhi",
    "country": "india"
}

## Iterate Thorugh keys 

## for key in dict : Loop through all keys 

print("Looping thorugh al keys")
for key in D:
    print(key)

## Iterate thorugh values
## loop through all values

print("Looping through Values only")
for value in D.values():
    print(value)

## Iterate through key-vlaue pairs 
## items()  : Iterate through both kew-value together

for key,value in D.items():
    print(key,":",value)


##Itearte using keys 
## .keys() : explicit loop thorugh keys

for key in D.keys():
    print(key)

# Note: for key in student: and for key in student.keys(): produce the same result. The first form is shorter and is used more often.




```

    Rule to Remember
    for key in dictionary:

    ➡️ Gives keys

    for value in dictionary.values():

    ➡️ Gives values

    for key, value in dictionary.items():

    ➡️ Gives both key and value

```python
# Student Record Management System

def student_record_management():

    student = {
        "name": "A",
        "age": 22,
        "course": "Data Science",
        "marks": 80
    }

    # 1. Display complete student record
    print("===== STUDENT RECORD =====")
    print(student)

    # 2. Add two new pieces of information
    student.update({
        "city": "Delhi",
        "email": "a@gmail.com"
    })

    # 3. Update existing information
    student["age"] = 21
    student["marks"] = 90

    print("\nUpdated Record:")
    print(student)

    # 4. Remove one piece of information
    student.pop("email")

    # 5. Check whether email exists
    if "email" in student:
        print("\nEmail Found")
    else:
        print("\nEmail Not Found")

    # 6. Safely retrieve city
    print(f"City : {student.get('city', 'Not Available')}")

    # 7. Display total pairs, keys and values
    print(f"\nTotal Key-Value Pairs : {len(student)}")
    print(f"Keys   : {list(student.keys())}")
    print(f"Values : {list(student.values())}")

    # 8. Display every key-value pair
    print("\nStudent Details:")

    for key, value in student.items():
        print(f"{key} : {value}")

    # 9. Create a backup
    backup = student.copy()

    # 10. Display keys alphabetically
    print("\nKeys (Alphabetical):")

    for key in sorted(student.keys()):
        print(key)

    # 11. Remove the last inserted key-value pair
    removed = student.popitem()

    print(f"\nRemoved Pair : {removed}")

    # 12. Display original and backup
    print("\nOriginal Dictionary:")
    print(student)

    print("\nBackup Dictionary:")
    print(backup)


student_record_management()
```
