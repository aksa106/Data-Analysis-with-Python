##### common operations 

```python
## concatentation

a = [1,2]
b = [3,4]

print(f"Concat {a+b}")

## repetation 

numbers = [1,2]

print(f"repetation : {numbers * 3}")

## len()

students = ["Rahul","Aman","Priya"]

print(len(students))
```

#### Search

```python
## Membership

## in/not in  : Checks whether an element is inside a list or not.

fruits = ["Apple","Banana","Mango"]

print("Banana" in fruits)

print("Orange" not in fruits)
```

```python
## index 
# list is ordered so every element has a fixed position, .index() returns the first occurrence of an element

fruits = ["Banana","Apple","Mango"]

print(fruits.index("Mango"))

## if an element does not exists ; list raises an value error
```

```python
## Count
# How many times an element appears
# Duplicate entries of an repeated value 

numbers = [10, 20, 10, 30, 10]

print(numbers.count(10))
```

#### Modify

```python
# Update Elements 

lst = ["A","B","C","D","E","F","G","H","I"]

lst[0] = "1" ## Update replaces an EXISTING element using its index

print(f"Updated {lst}")

# Add Elements 

# append : ADD ONE ELEMENT AT THE END OF A LIST.  ("")

lst.append("J")

print(f"Appended : {lst}")

# Extend : ADDS MULTIPLE ELEMENTS TO THE END OF LIST  ([])

lst.extend(["K","L","M","N"])

print(f"Extended : {lst}")

# Insert : INSERT ELEMENT AT AN SPECIFIC INDEX

lst.insert(1,"First Element")

print(f"Inserted : {lst}")

## Remove : remove the first occurence of a value.

lst.remove("First Element")

print(f"Remove : {lst}")

## pop : Remove and returns an element by index , removes from end

lst.pop()

## clear() :removes all elements from list

lst.clear()

print(lst)
```

```python
## Arrange Elements 


# sort()

number = ["c","b","a"]

num = [30,20,10]

number.sort()
num.sort()

num.reverse()
 
num1 = num.copy() ## created a shallow copy of the list

print(f"number sorted : {number} \nnum sorted : {num} \nReverse{num}")
print(num1)
```

#### Slicing

```python
lst = [1,2,3,4,4,5]

print(lst[:])

print(lst[::-1])

print(lst[:3])

print(lst[3:5])
```

#### Iteration

```python
## for loop 

lst = ["a","b","c","d","e","F"]

for i in lst:
    print(i)
```

```python
# while 
lst = ["a","b","c","d","e","F"]

i = 0
while i<len(lst):
    
    print(i,lst[i])
    i+=1


```

```python
## range(len) : when u need index as well as value 

lst = ["a","b","c","d","e","F"]

for i in range(len(lst)):
    print(i,lst[i])
```

```python
## enumerate()

## enumerate gives both index and eleemt ditectly, its cleaner than range

lst = ["a","b","c","d","e","F"]

for index,lst in enumerate(lst):
    print(index,lst)
```

```python
## Practice 1 : Student Marks Manager 

def student_marks():
    lst = [50,60,70,80,90]

    print("All marks")
    for i in lst:
        print(i)

    print(f"First Mark : {lst[0]}")
    print(f"Last Mark : {lst[-1]}")


    print("Total number of marks")
    sum = 0
    for i in lst:
        sum += i

    print(sum)


    ##Adding a new mark
    lst.append(100)
    print(f"Added one more mark : {lst}")

    ##Removing one mark
    lst.remove(50)
    print(f"Removed one mark : {lst}")

    print(f"Updated Final list {lst}")

    


student_marks()
```

```python
## Shopping Cart 

def shopping_cart():

    cart = ["Cleanser","Toner","serum","Cream","Essence"]

    i=0
    while i < len(cart):
        print(cart[i])
        i+=1


    
    cart.extend(["BodyWash","Perfume"])
    print(f"new cart : {cart}")

    cart.insert(2,"Roll on")
    print(cart)

    
    print(cart.pop())
    print(cart)


    print(f"First 3 products : {cart[:3]}")
    print(f"Last 2 products : {cart[-2:]}")
    print(f"Reversed : {cart[::-1]}")

    

shopping_cart()

```

```python
## Library Book Manager

def library_manager():

    books = ["Alchemist","Vernoica","Zahir","48 Laws","Atomic habits"]

    print("Current Collection : ")
    for book in books:
        print(book)

    books.extend(["Python","Practical Stats for DS"])
    print(f"Updated Collection : {books}")

    books.remove("Python")
    print(f"{books}")

    books.insert(0,"Clean code")
    print(f"{books}")

    print(f"First book : {books[0]}")
    print(f"Last book : {books[-1]}")
    print(f"First Four books : {books[0:4]}")
    print(f"Every second book : {books[::2]}")
    print(f"Reverse : {books[::-1]}")



    #how many books are currnetly in library
    total_count = len(books)
    print(total_count)

    #Check if a book exists

    print("Clean code" in books)
    

    ## Arranging the collection alphabetically

    books.sort()
    print(books)


    backup = books.copy()

    print(backup)



library_manager()

```
