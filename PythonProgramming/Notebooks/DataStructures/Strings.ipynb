# ACCESSING DATA

### INDEXING

* Positive indexing : starts from beginning (0,1,2 ..)

* Negative indexing : starts from end (-1,-2,-3 ..)

* index error : occurs when u try to accces an idex that doesnt exists

```python
text = "PythonProgramming"

Positive_index = text[6]

Negative_index = text[-6]

print(Positive_index)

print(Negative_index)
```

#### Slicing 

* Extract a portion (substring) of a string.

* (start, stop, step) : stop : excluded [0:10:2]

* deafult : if u dont provide start stop step by def python uses start and end of str

* Reverse Slicing ([::-1]) → Creates a reversed copy of the string.

```python
# Slcing 

text = "PythonProgramming"

## Extract Python 

Python = text[0:6]
print(Python)
## Extract Program 
program = text[6:13]

print(program)

## Reverse 

print(text[::-1])

##negative index

print(text[-1:-12])

```

#### String length 

    len() → Returns the total number of characters in a string.
    Spaces Count → Spaces are counted as characters.
    Special Characters Count → Symbols like @, #, _ are also counted.
    Empty String → len("") returns 0.

# Search

```python
## in /not in : checks whether a substring exists in a string or not

text = "Python"

print("Py" in text)

```

```python
## FINDING CHARCHTERS / SUBSTRINGS 

# find() : finds the first occurence of a char index [returns -1 if not found]

text = "PythonProgramming"

print(text.find("Py"))

print(text.find("ja"))  ## -1 not found 

#----------------------------------------------------------

## index() : finds index of first occurence of char , raises ValueError if not found

print(text.index("Pro"))

print(text.index("Ja"))
```

```python
## count()

text = "PythonProgramming"

print(text.count("P"))
```

###  Prefix and Suffix Checking 

```python
## PREFIX (startswith())

text = "PythonProgramming"

print(text.startswith("P"))

## SUFFIX (endswith)

print(text.endswith("g"))
```

# Modify
#### Strings are immutable, so these methods do not change the original string. They return a new string.

```python
## CASE CONVERSIONS & TEXT methods

text = "PythonProgramming"

print(text.upper())

print(text.lower())

print(text.title())

print(text.capitalize())

print(text.replace("Python","Java"))



print(text.strip())
print(text.lstrip())
print(text.rstrip())


print(len(text))
```

```python
## SPLIT BREAKS A STRING INTO A LIST USING A SEPARATOR

text = "Python SQL Excel"

print(text.split())

## Custom separatpr 

text1 = "Python SQL"

print(text1.split(","))
```

```python
## JOIN - joins multiple strings together CONVERTS LIST TO A STR

items = ["A","B","C"]

print(",".join(items))
```
