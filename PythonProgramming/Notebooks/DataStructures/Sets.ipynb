```python
def employee_analyzer():

    while True:

        employee_id = input("Enter Employee ID: ").strip().lower()

        if len(employee_id) >= 8:
            break

        print("Invalid Employee ID")

    print("\n===== EMPLOYEE ID ANALYSIS =====")

    # Access
    print(f"Employee ID          : {employee_id}")
    print(f"Length               : {len(employee_id)}")
    print(f"First Character      : {employee_id[0]}")
    print(f"Last Character       : {employee_id[-1]}")
    print(f"First 3 Characters   : {employee_id[:3]}")
    print(f"Last 4 Characters    : {employee_id[-4:]}")
    print(f"Every 2nd Character  : {employee_id[::2]}")
    print(f"Reverse              : {employee_id[::-1]}")

    # Search
    print(f"\nContains 'python'    : {'python' in employee_id}")
    print(f"Starts With 'emp'    : {employee_id.startswith('emp')}")
    print(f"Ends With '2026'     : {employee_id.endswith('2026')}")
    print(f"Count of 'o'         : {employee_id.count('o')}")

    # Loop
    print("\nCharacters:")

    for ch in employee_id:
        print(ch)


employee_analyzer()
```
