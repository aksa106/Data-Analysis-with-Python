```python
# user information

username = "Rahul"
password = "python123"


#Login details entered by the user 
entered_username = "Rahul"
entered_password = "python123"


##Check if user info matches with login details 

is_username_correct  = entered_username == username 
is_password_correct = entered_password == password

print(is_password_correct)
print(is_username_correct)


## DEcision  based on boolean values 

if (is_username_correct and is_password_correct):   ## if () -> True
    print("Login successfull")
else:
    print("invalid username or password")

```

### (if-if-if) : Independent checks 
* When multiple conditions can be true and you want to excute all the blocks 
* Each if is independent , python chekcs for every if.

### (if-elif-elif-else) : Decision chain 
* As soon as one condition is True python stops checking.
* Even though other consitions in elifs can be true , but as soon as condition becomes True Python stops checking, Python stops checking after the first matching condition.


# if-if-if : Wants to execute all matching blocks based on all true conditions.
# if-elif-elif : when only one outcome should happen.
