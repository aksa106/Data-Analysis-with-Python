```python
# Boolean Values

## Boolean is data type in Python which has two values True and False

#Comparison operations produce Boolean values.
```

```python
# user information

username = "Rahul"
password = "python123"


#Login details entered by the user 
entered_username = "Rahul"
entered_password = "python123"


##Check if user info matches with login details 

is_username_correct  = entered_username == username 
is_password_correct = entered_password == password

print(is_password_correct)
print(is_username_correct)


## DEcision  based on boolean values 

if (is_username_correct and is_password_correct):   ## if () -> True
    print("Login successfull")
else:
    print("invalid username or password")

```
