# Loops (for & while)


* Purpose : for loop iterates over a sequence and collection where , while loop iterates until the given condition is True.

* Best Used When 
    * for loop : we know what we are iterating over and how many iterations do we need.
    * while loop : we don't know how many iteraions and it depends on a condition.

* Works with
    * for loops : always works with an iterable and collections(list, str, tuple, dict)
    * while loops : Boolean conditions (can also work with collections but not recommended)

* Requires index : Only while loops 

* Manual update required : only in while loops (i+=1)

* Supports break : both

* supports continue : both

* supports pass : both 

* suports else : both 

* REAL WORLD EXAMPLE : 
    * for : displays users, products , results, orders, products, process files, iterate over database results.

    * while : Login attempts, input validation, menu-driven programs, retry until success

# Mental Model 
* print() : loop -> print() -> Continue Loop -> Next iteration (function stays alive) print just have only one job keep displaying the outuo , whether output is single or multiple in iteration.
##### Show every iteration - use print()


* return ()
    * Loop -> return -> Fucntion exit

    * return exits the fucntion immediately.

    * code written after return is unreachable..

    * A fucntion or any code stops after its first return

    * Return can return any data type

    * Functions without return returns none.

    * Return does not prints anything, its print() job to display the output.

    * return is how functions communicate

    * Every function should have a clear purpose


* return immediately exits the fucntion, code after return never runs, return does not print anything, without return python does not print anything and it returns none, Put return after the loop if you want to process every iteration before returning.


##### Stop immediately and send one result : return inside loop

##### Process All iterations and send final result : return after loop 




```python
# Iterate over a list

def display_users():

    users = ["A","B","C"]

    for user in users:
        print(user)

display_users()
```

```python
# Iterate over a string

def display_charchter():
    string = "PythonProgramming"

    for char in string:
        print(char)

display_charchter()
```

```python
# Using range (stop)

def count_to_four():
    for i in range(5): #5=excluded
        print(i)

count_to_four()
```

```python
# Using range (start, stop)

def count_to_5():
    for i in range(2,5):
        print(i)

count_to_5()
```

```python
# Using range(start,stop,step)

def even_no():
    for i in range(2,10,2):
        print(i)

even_no()
```

## while

```python
## countdown

def countdown():
    number = 5 

    while number >= 1:
        print(number)
        number = number - 1 

    print("Blast OFF!")

countdown()

```

```python
## Input Validation 

def login():
     
     password = ""

     while password != "python123":
          
          password = input("Enter Password : ")

     print("Login Sucessful")

login()



```
