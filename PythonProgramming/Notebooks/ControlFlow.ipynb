# Loops (for & while)


* Purpose : for loop iterates over a sequence and collection where , while loop iterates until the given condition is True.

* Best Used When 
    * for loop : we know what we are iterating over and how many iterations do we need.
    * while loop : we don't know how many iteraions and it depends on a condition.

* Works with
    * for loops : always works with an iterable and collections(list, str, tuple, dict)
    * while loops : Boolean conditions (can also work with collections but not recommended)

* Requires index : Only while loops 

* Manual update required : only in while loops (i+=1)

* Supports break : both

* supports continue : both

* supports pass : both 

* suports else : both 

* REAL WORLD EXAMPLE : 
    * for : displays users, products , results, orders, products, process files, iterate over database results.

    * while : Login attempts, input validation, menu-driven programs, retry until success

```python
# Iterate over a list

def display_users():

    users = ["A","B","C"]

    for user in users:
        print(user)

display_users()
```

```python
# Iterate over a string

def display_charchter():
    string = "PythonProgramming"

    for char in string:
        print(char)

display_charchter()
```

```python
# Using range (stop)

def count_to_four():
    for i in range(5): #5=excluded
        print(i)

count_to_four()
```

```python
# Using range (start, stop)

def count_to_5():
    for i in range(2,5):
        print(i)

count_to_5()
```

```python
# Using range(start,stop,step)

def even_no():
    for i in range(2,10,2):
        print(i)

even_no()
```

## while

```python
## countdown

def countdown():
    number = 5 

    while number >= 1:
        print(number)
        number = number - 1 

    print("Blast OFF!")

countdown()

```

```python
## Input Validation 

def login():
     
     password = ""

     while password != "python123":
          
          password = input("Enter Password : ")

     print("Login Sucessful")

login()



```

### break immediately exits the loop , unlike return, it does not exit the function , it exits only the loop

```python
## Example : search for a user (for loop)

def find_users():
    users = ["Rahul","Priya","Arun"]

    for user in users: 
        print(f"Checking {user}")

        if user == "Priya":
            print("user found")

            break 

    print("Search completed")

find_users()



```

## continue  : does not stops the loop , It skips the current iteration and moves to the next one

```python
# Example : Skip the admin user (for loop)

def display_users():
    users = ["Rahul","Admin","Arun"]


    for user in users:
        if user == "Admin":
            continue 
        print(user)

    print("All user processed")

display_users()


## Admin was not printed , the loop did not stop, it simply skipped the whole iteration 

```

```python
## Skip even numbers {for loop}



def odd_numbers():
    for i in range(1,11):
        if i % 2 == 0:
            continue 
        print(i)

    print("Odd numbers")

odd_numbers()
```

```python
def odd_number():
    num = 0 

    while num < 10:
        num += 1

        if num % 2 == 0 :
            continue
            
        print(num)

    print("odd numbers")

odd_number()

```

### pass : pass does nothing.

It is simply a placeholder that tells Python:

"There's no code here yet, but don't give me an error."

```python
## Placeholder in a  loop 

def process_users():

    users = ["Rahul","Arun","Admin"]

    for user in users:

        if user == "Rahul":
            pass 

        print(user)


process_users()



```

# break Vs Continue Vs pass 

* break : Exits the loops , does not exits the fucntion 

* continue : skips current iteration ,  loop exits and function doesnt ends

* pass : does nothing, loop function , fucntion does not ends 

* return : completely ends the loop and exits fucntion




```python
def sum_using_for(n):

    total = 0

    for i in range(1, n + 1):
        total += i

    print(total)

sum_using_for(5)
```
