# PANDAS - DATAFRAME & SERIES 

Pandas is a poweful data manipulation library in Python widely used for data cleaning and data analysis.
It provides two primary data structures 

* Series :  One dimensional array-like object 

* Data Frames : Two Dimensional , size-mutable, and potentially hetrogeneous tabular data structure with labelled axes (rows & columns).






```python
#A Series
#It is a one dimensional array like object , that can hold onto any data type. It is similar to a column in a table.

import pandas as pd
data = [1,2,3,4,5]
series = pd.Series(data)
print("Series \n", series)
print(type(series))
```

```python
# Create a Series from dictionary
data = {"a":1,"b":2,"c":3}     
# defualt index in dict was - 0 , 1, 2
# now my index becomes - a, b, c
series_Dict = pd.Series(data)
print(series_Dict)
```

```python
data = [10,20,30]
index = ["a","b","c"]
pd.Series(data,index=index)
```

# DATA FRAME

* A data frame in pandas is basically just like a table, just like an excel sheet
* It contains rows and columns
* It is mutable
* It is heterogeneous - can hold multiple data types 


```python
## Data Frame 
## How to create a data frame completely from basics
## from a dictionary of list


data = {
    "Name" : ["A","B","C"],
    "Age" : [25,30,45],
    "city" : ["Bangalore","New york","Delhi"]
}
## 3 columns , Many rows 

df = pd.DataFrame(data)
print(f"{df} \n {type(df)}")
```

```python
import numpy as np
np.array(df)
```

```python
## Create a Data frame from a list of dict

data1 = [
    {"Name" :"A", "Age":25, "City" : "Delhi"},
    {"Name" :"B", "Age":45, "City" : "Punjab"},
    {"Name" :"C", "Age":20, "City" : "Himachal"},
    {"Name" :"D", "Age":25, "City" : "Haryana"},
]
df = pd.DataFrame(data1)

print(f"{df}\n {type(df)}")
```
