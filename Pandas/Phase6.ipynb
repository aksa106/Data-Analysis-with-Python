DATA FRAME INFORMATION AND SUMMARY THOERY

```python
import pandas as pd

data = {
    "Name": ["A", "B", "C"],
    "Age": [21, 22, 23],
    "Salary": [50000, 60000, 70000]
}

df = pd.DataFrame(data)

print(df)
```

### shape

```python
print(df.shape) ## 3 rows 3 columns
```

### dtypes

```python
print(df.dtypes)

## text = object
## number = int64
```

### info()

```python
print(df.info())

## dataset structure
## columns
## datatype
## non null values
```

### Describe

```python
print(df.describe())

## gives statistical summary
# count, mean, std, min, max

#works mainly on numeric cols like -- Age, Salary
```
