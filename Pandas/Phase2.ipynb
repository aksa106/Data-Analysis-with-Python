### CREATING DATA

```python
##CREATING SERIES FROM LIST

import pandas as pd

s = pd.Series([10,20,30,40])

print(s)
```

```python
##CREATING SERIES FROM A DICTIONARY

data = {
    "a" : 10, 
    "b" : 20,
    "c" : 30
}

s = pd.Series(data)

print(f"Series (created from dict)  \n{s}")

print(f"Indexes \n{s.index}")
```

```python
# CREATE DATA FRAME FROM DICTIONARY (Of lists)

data = {
    "Name" : ["A","B","C"],
    "Age"  : [21,22,23]
}

df = pd.DataFrame(data)

print(f"Data Frame \n{df}")

## Keys -> Column names
## Lists -> Column data

print(f"Index : \n{df.index}")
```

```python
# CREATE A DATA FRAME FROM LIST OF DICTIONRIES

data = [
    {"Name":"A", "Age":21},
    {"Name":"B", "Age":22},
    {"Name":"C", "Age":23}
]

## EACH DICT = ONE ROW

df = pd.DataFrame(data)

print(df)

print(f"Columns \n{df.columns}")
```
