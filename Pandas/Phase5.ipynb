### DATA MANIPULATION

* Changing data
* Updating data
* Removing data
* Adding data

```python
import pandas as pd 

data = {
    "Name" : ["A","B","C"],
    "Age" : [21,22,23]
}

df = pd.DataFrame(data)

print(df)
```

### ADD NEW COLUMN

```python
## add a new column salary

df["Salary"] = [5000,6000,7000]

print(df)
```

### UPDATE EXISTING VALUES

```python
df["Age"] = df["Age"] + 10

print(df)
```

### DROP A COLUMN

```python
## TEMPERORY DROP

df.drop("Salary",axis=1)

## axis = 1 means
## Column operation is not permanent yet
```

```python
## PERMANENT DROP

df.drop("Salary", axis=1, inplace = True)

# now permanently removed

print(df)
```

```python
df.drop(0, inplace=True)

print(df)
```

```python
df.drop(2, inplace=True)
print(df)
```
