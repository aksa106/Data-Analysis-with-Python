### ACCESSING DATA

```python
import pandas as pd 

data = {

    "Name" : ["A","B","C"],
    "Age"  : [21,22,23],
    "City" : ["Delhi", "Mumbai", "Pune"]
}

df = pd.DataFrame(data)

print(df)
```

### ACCESS ONE COLUMN

```python
print(f"Accessing First Column : \n{df["Name"]}")
```

### LOC VS ILOC

* LOC - labels/names, column names, row labels/indexes, 

* ILOC - numeric positions, column indexes, row positions

```python
print(df)
print("\n")

##-----------------------------------------------------

print(f"LOC (Third column(city),second index ) \n{df.loc[1,"City"]}")

print("\n")

print(f"ILOC (Third column(city),second index ) \n{df.iloc[1,2]}")
```

```python
import pandas as pd 

data = {

    "Name" : ["A","B","C"],
    "Age"  : [21,22,23],
    "City" : ["Delhi", "Mumbai", "Pune"]
}

df = pd.DataFrame(data)


print(f"LOC (All rows of a specific column[age])          \n{df.loc[:, "Age"]}")

print("\n")

print(f"(All rows of a specific column[age])               \n{df.iloc[:,1]}")

print(df["Age"])

print(df.loc[0:1,"Age"])

print(f"(All rows of a specific column[age])               \n{df.iloc[0:1,0:2]}")
```

```python

print(f"LOC (Entire First Row : \n{df.loc[0]})")

print("\n")

print(f"ILOC (Entire First Row : \n{df.iloc[0]})")
```

### AT VS IAT

* Both are used for accessing one values only
* at = lables/names
* iat = numeric positions

* Exactly same logic as:

    * loc
iloc

but ONLY for one cell.

```python
import pandas as pd

data = {

    "Name" : ["D","E","F"],
    "Age"  : [21,22,23],
    "City" : ["Delhi", "Mumbai", "Pune"]
}

df = pd.DataFrame(data)
print(df)
```

```python
print(f"AT (Row 1 and Column(City)) \n{df.at[1,"City"]}")
```

```python
print(f"IAT  (Row1 column1) \n{df.iat[1,2]}")
```

### LOC   VS     AT

* [LOC] is flexible it can return 
    * One value
    * Full rows
    * Multiple Values of a specifc row/col

* [AT]  is strict -- 
    * it ONLY accesses one exact single value/cell
    * using row/column lables

```python
## OBSERVATION 1

# We tried to acces only one value which is First row in column City and we got it because , loc and at can both access on value


print(f"AT (First City) : \n{df.at[0,"City"]} ")

print("\n")

print(f"LOC (First City) : \n{df.loc[0,"City"]}")
```

* loc and at can both access data using row labels and Column names, they cna give the same output for ONE CELL.

```python
## OBSERVATION 2

print(f"LOC (Entire First Row) : \n{df.loc[1]}")

## loc can acces entire Row 
```

```python
print(f"AT(Entire First Row using AT : )")

print(df.at[1])

# at cannot access entire row as it is used for accessing only ONE value.
```

```python
## OBSERVATION 3

print(f"LOC(All rows from Age Column) : \n{df.loc[:,"Age"]}")

#Loc can access all of the Rows of a specific column
```

```python
print("All rows of age column using AT")

print(df.at[:,"Age"])

# at cannot access all of the rows since it can only access ONLY ONE value
```

### IAT VS ILOC

    * iloc and iat can both access data using numeric row/column positions, so they can return the same value.

    * iloc is flexible — it can return rows, columns, or multiple values.
    
    * iat is strict — it ONLY returns one exact single cell value.

```python
import pandas as pd 

data = {

    "Name" : ["A","B","C"],
    "Age"  : [21,22,23],
    "City" : ["Delhi", "Mumbai", "Pune"]
}

df = pd.DataFrame(data)

print(df)
```

```python
print(f"ILOC  (Access mumbai) : \n{df.iloc[1,2]}")

print("\n")

print(f"ILOC (Access Full ROW)  \n{df.iloc[2]}")
```

```python
print(f"IAT (Access mumbai)  \n{df.iat[1,2]}")
```

```python
print("Access full row using iat")

print(df.iat[2])  ##iat fails at this
```
