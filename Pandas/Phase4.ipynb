### ACCESSING DATA

```python
import pandas as pd 

data = {

    "Name" : ["A","B","C"],
    "Age"  : [21,22,23],
    "City" : ["Delhi", "Mumbai", "Pune"]
}

df = pd.DataFrame(data)

print(df)
```

### ACCESS ONE COLUMN

```python
print(f"Accessing First Column : \n{df["Name"]}")
```

### LOC VS ILOC

* LOC - labels/names, column names, row labels/indexes, 

* ILOC - numeric positions, column indexes, row positions

```python
print(df)
print("\n")

##-----------------------------------------------------

print(f"LOC (Third column(city),second index ) \n{df.loc[1,"City"]}")

print("\n")

print(f"ILOC (Third column(city),second index ) \n{df.iloc[1,2]}")
```

```python
print(f"LOC (All rows of a specific column[age])          \n{df.loc[:, "Age"]}")

print("\n")

print(f"(All rows of a specific column[age])               \n{df.iloc[:,1]}")
```

```python

print(f"LOC (Entire First Row : \n{df.loc[0]})")

print("\n")

print(f"ILOC (Entire First Row : \n{df.iloc[0]})")
```

### BIGGEST Mental Difference
* LOC

    you think in names
    * "Age"
    * "City"

* ILOC

    You THINK in positions
    * 0
    * 1
    * 2

### Most real code uses : loc

### AT VS IAT

* Both are used for accessing one values only
* at = lables/names
* iat = numeric positions

* Exactly same logic as:

    * loc
iloc

but ONLY for one cell.
