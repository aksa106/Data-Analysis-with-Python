# JSON OPERATIONS 


#### what is json?
* Json = Javascript Object Notation
    * a lightweight data format
    * used to store and exchange data

* JSON STORES DATA AS
    * {
        "key" : "Value"
    }

    * Very common in APIs , WebApps , Backend systems

```python
## What problem JSON solves 

## imagine this in python

name = "Aksa"

age = 22

role = "Data Analyst"

```

### Now suppose :
* website wants data
* backend wants data
* API wants data
* ------------------------- So sending 3 separate variables is kinda messy (So we package them together)

## JSON is just a PACKET of data

```python
## Like this :

{
    "name" : "Aksa",
    "age" : 22,
    "role" : "Data Analyst"
}

## THIS IS JSON


##-------------- JSON STRUCTURE -----------------------

# left side = "name" -- mainly KEYS
# right side = "Aksa" -- Values

# Label : actual data
```

#### Why industry loves JSON
* Because:
    * humans and machines can read it
    * works eveywhere

* Used in :
    * APIs
    * WebApps
    * MobileApps
    * DB
    * AI systems

### JSON IS NOT PANDAS 

### IT IS SIMPLY A DATA FORMAT pandas simply reads it.

* Excel file → Pandas reads
* CSV file → Pandas reads
* JSON file → Pandas reads

* JSON = container of data
* Pandas = tool that reads container
