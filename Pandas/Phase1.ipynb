#### PANDAS

```python
import pandas as pd

## Creating a series

series = pd.Series([21,22,23])

print(f"[Series]  \n{series}")


print("\n")
#Values at idx
print(f"First element : {series[0]}")
print(f"Second element : {series[1]}")
print(f"Third element : {series[2]}")


print("\n")
#Datatype
print(f"Type : {type(series)}")

print("\n")

#Check index
print(f"Index : {series.index}")
```

```python
## CREATING A DATA FRAME

data = {
    "Name" : ["A","B","C"],
    "Age" : [21,22,23]
}

df = pd.DataFrame(data)

print(f"[ Data Frame]   \n{df}")

print("\n")
#----------------------------------------------------------

# Columns - Name, Age
# Rows   - A,21
# Index  - 0,1,2


#   THIS IS A TABLE -> DATA FRAME

## -----------------------------------------------

#check index
print(f"Index : {df.index}") # indexes go from 0-2

print("\n")

#Check column names
print(f"Column names : {df.columns}")

print(f"Rows : {df.shape}")

print("\n")

#specific row
print(f"Total Rows : {df.shape[0]}")
```
