# CATEGORICAL AND DISTRIBUTION PLOTS

    Categorical Plots

        Bar Plot → compare category averages
        Box Plot → median, spread, outliers
        Violin Plot → distribution shape + density

    Distribution Plots
    
        Histogram → frequency distribution
        KDE Plot → smooth distribution curve

## BAR PLOT

```python
# Bar Plot = Categorical vs Numerical
```

```python
## Find out on which day is the avergae resturant bill is highest?
# X = day
# Y = total_bill

import seaborn as sns , matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.barplot(
    x="day",
    y="total_bill",
    data=tips
)

plt.title("Total Bill by Day")

plt.show()
```

## What Seaborn is secretly doing

Seaborn is not plotting every bill

    It groups by day (Thurs,Fri,Sat,Sun) 
    Then it calculates mean total_bill for each day 

        Conceptually : 
        tips.groupby("day")["total_bill"].mean()
        Then it plots the average

## BOX PLOT (Imp)

## What it shows 
* Median
* Spread of data
* Outliers
* Quartiles (25%,50%,75%)

```python
import matplotlib.pyplot as plt , seaborn as sns

tips = sns.load_dataset("tips")

sns.boxplot(
    x="day",
    y="total_bill",
    data=tips
)

plt.show()
```

    Bottom of the box = 25th percentile (Q1)

    Middle line = Median (Q2)

    Top of the Box = 75th percentile (Q3)

    Dots Outside = Outliers 



    Which day seems to have the most outliers?
    Thur

    Which day appears to have the highest median total bill
    sun


    1. Most outliers?

    I'd say Saturday has the most obvious outliers.

    Count the dots:

    Thur ≈ 5 outliers
    Fri ≈ 1 outlier
    Sat ≈ 3 outliers
    Sun ≈ 2 outliers

    However, Saturday's outliers are more extreme (higher values).

    So your answer isn't crazy, but if someone asked me "which has the most?", I'd pick Thursday based on count.

    2. Highest median?

    ✅ Sunday

    The median is the horizontal line inside the box, and Sunday's is the highest.


## VIOLIN PLOT (used occasionally)

* Unlike the box plot, a voilin plot shows the shape of the distribution

* wider sections = more data points are there,


```python
import matplotlib.pyplot as plt , seaborn as sns

tips = sns.load_dataset("tips")

sns.violinplot(
    x="day",
    y="total_bill",
    data=tips
)

plt.show()
```

* Wider parts = lots of bills around that value

* Narrow parts = fewer bills around the value

    A box plot tells you:

    median
    quartiles
    outliers

    A violin plot tells you:

    all of the above plus
    where the data is concentrated.

    For example, for Sun and Sat, you can see the widest region is roughly around 15–20 total_bill, meaning many observations are clustered there.


## HISTOGRAM

```python
import seaborn as sns, matplotlib.pyplot as plt 

tips = sns.load_dataset("tips")

sns.histplot(
    tips["total_bill"],  
    ## This is the column being analyzed, we are asking "Show me the distribution of resturant bills (Not tips, days, [Only bills])"


    bins=10,
    # Bin = A range
    ## suppose values goes from 3 to 50
    # with 10 bins seaborn splits that range into 10 intervals


    kde=True
    ## with kde seaborn draws a smooth curve over the bars
    ## The curve tries to show the overall shape of the distribution


)

plt.show()

## THE BAR HEIGHT IS -- COUNT
```

### Histogram
* How frequently do different values occur

    Suppose bills are :
    10, 12, 15, 18, 19, 21, 22, 24, 50

    Instead of showing every number individually, a histogram groups them into ranges

    0-10  | *
    10-20 | *****
    20-30 | 

    Now you can instanly see where does most data lives

------------------------------------------------------------------------------------

    For a right-skewed distribution:

    Mean > Median

    For a left-skewed distribution:

    Mean < Median

    For a symmetric distribution:

    Mean ≈ Median


## KDE PLOT (Kernel Density Estimation)

```python
import matplotlib.pyplot as plt , seaborn as sns

tips = sns.load_dataset("tips")

sns.kdeplot(
    tips["total_bill"],
    fill = True
)

plt.show()
```

### A KDE plot is basically a smooth version of histogram 

Instead of buckets/bins it estimates a continous probability density curve 

Histogram can look different depending on bins, so KDE tries to reveal the underlying distribution without depending on bin boundries
