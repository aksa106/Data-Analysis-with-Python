# Scatter Plot (First Real Seaborn Plot)

* From the dataset tips:
    X = total_bill
    Y = tip

    Q: "As the bill increases, does the tip tend to increase??
    That is what scatter plots are for, To find relationship between two numerical variables, Pattern detection, Correlation intuition, Outlier spotting

### Practical Task

    Create a scatter plot using:

    x = total_bill
    y = tip
    data = tips

    Then observe the graph and answer:

    Does there appear to be a positive relationship between bill and tip?
    Are the points tightly clustered or fairly spread out?


```python
import seaborn as sns , matplotlib.pyplot as plt 

tips = sns.load_dataset("tips")

sns.scatterplot(
    x = "total_bill",
    y = "tip",
    data = tips
)

plt.title("Total Bill vs tip")

plt.show()
```

## What did this plot teach us?

    We are seeing a positive trend, 
    As total_Bill increases, tip genrally increaes



## LINE PLOT





```python
import seaborn as sns , matplotlib.pyplot as plt 

tips = sns.load_dataset("tips")

sns.lineplot(
    x="size",
    y="total_bill",
    data = tips
)

plt.title("Total Bill by Size")

plt.show()
```

The graph looks strange because a line plot is usually used for ordered data (time series, sequence, trends).

    Here:

    size = number of people at the table (1, 2, 3, 4...)
    Seaborn is automatically calculating the average total_bill for each size and connecting them with a line.

    So you're not seeing individual observations like in the scatter plot. You're seeing an aggregated trend.
