# CUSTOMIZATION AND REAL DATASET

## Part A -- Plot Customization 

* Changing how the graph looks, not what data it shows


### Figure Size

```python
import matplotlib.pyplot as plt, seaborn as sns

tips = sns.load_dataset("tips")

sns.barplot(
    x="day",
    y="total_bill",
    data=tips
)

plt.show()

##------------------------------------------

# Default , matplotlib decides the graph size, 
```

```python
## Custom Size

import matplotlib.pyplot as plt, seaborn as sns

tips = sns.load_dataset("tips")

plt.figure(figsize=(10,6))   ## Custom width & height 10,6
 
sns.barplot(
    x="day",
    y="total_bill",
    data=tips
)

plt.show()
```

```python
plt.figure(figsize=(4,3))

sns.barplot(
    x="day",
    y="total_bill",
    data=tips
)

plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.barplot(
    x="day",
    y="total_bill",
    data=tips
)

plt.show()
```

## TITLE

```python

import matplotlib.pyplot as plt, seaborn as sns

tips = sns.load_dataset("tips")



sns.barplot(
    x="day",
    y="total_bill",
    data=tips
)

plt.title("Total Bill by Day")

plt.show()
```

## AXIS LABELS

```python
import matplotlib.pyplot as plt , seaborn as sns

tips = sns.load_dataset("tips")

sns.barplot(
    x="day",
    y="total_bill",  ## Seaborn automatically uses total_bill as label, but these labels can become ugly and technical

    data=tips
)


## by adding these graph became more human readable

plt.title("Total Bill by Day")

plt.xlabel("DAys")

plt.ylabel("Total Bill ($)")

plt.show()
```
