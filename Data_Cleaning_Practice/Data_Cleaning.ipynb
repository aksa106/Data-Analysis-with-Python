```python
## Category Validation — Consistency & Standardization

categorical_cols = [
        'Customer_Name',
    'Gender',
    'State',
    'Product',
    'Category',
    'Payment_Mode'
]

for col in categorical_cols:
    print(f"\n{col}")
    print(df[col].unique())
```

```python
df['Gender'] = df['Gender'].replace(
    {
        'M' : 'Male',
        'F' : 'Female'
    }
)
```

```python
df['State'] = df['State'].replace(
    {
        'Maharastra' : 'Maharashtra'
    }
)
```

```python
categorical_cols = [
        'Customer_Name',
    'Gender',
    'State',
    'Product',
    'Category',
    'Payment_Mode'
]

for col in categorical_cols:
    print(f"\n{col}")
    print(df[col].unique())
```
