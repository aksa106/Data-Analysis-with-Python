```python
# 1. Create a Pandas DataFrame with 3 columns and 5 rows filled with random integers. Introduce some NaN values. Fill the NaN values with the mean of the respective columns.

import pandas as pd 

data = {
    "A":[1,2,3,4,5],
    "B":[10,11,12,13,14],
    "C":[20,21,22,23,24]
}

df = pd.DataFrame(data)
#Introduce some NaN values.

df.loc[:2,"A"] = None

df.loc[:3,"C"] = None

print(df)
#--------------------------------------------------

## inspect missing values in df 

missValues = df.isnull()

print(missValues)

missValues_count = df.isnull().sum()

print(missValues_count)

#------------------------------------------------------------

##Filled missing values

df_new = df.fillna(df.mean())

print(df_new)
```

## dropna()

* dropna() removes missing values (NaN) from a DataFrame.
    * dropna()



* Drop columns containing NaN
    * dropna(axis=1)

* Drop rows only if ALL values are NaN
    * dropna(how="all")

* Drop rows if ANY value is NaN
    * df.dropna(how="any")

* Keep changes permanently
    * df.dropna(inplace=True)

```python
# 2. Create a Pandas DataFrame with 4 columns and 6 rows filled with random integers. Introduce some NaN values. Drop the rows with any NaN values.

import pandas as pd 

data = {
    "A":[1,2,3,4,5],
    "B":[10,11,12,13,14],
    "C":[20,21,22,23,24]
}

df = pd.DataFrame(data)


df.loc[:2,"A"] = None

print(df)

#--------------------------------------------



df_new = df


df_new = df_new.dropna()

print(df_new)

```
