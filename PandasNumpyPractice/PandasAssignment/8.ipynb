## pivot()

* A summary table that groups data and calculate statistics.
* It is so hard to see patterns immediately so a pivot table organises it.

    A pivot table answers questions like:

    Total sales by product?
    Average salary by department?
    Revenue by year and quarter?
    Orders by city and month?

        Raw Data
        |
        Group
        |
        Calculate(sum/mean/count)
        |
        Display as summary table


        pd.pivot_table(
            df,
            values = What to calculate (numerical col)
            index = row labels,
            columns = Column labels
            aggfunc =  How to calculate (sum, mean, count, max, min)
        )



        

```python
# 1. Create a Pandas DataFrame with columns 'Date', 'Category', and 'Value'. Create a pivot table to compute the sum of 'Value' for each 'Category' by 'Date'.

import pandas as pd

data = {
    "Date": ["2026-06-06", "2026-06-06", "2026-06-07", "2026-06-07"],
    "Category": ["Phone", "Laptop", "Phone", "Laptop"],
    "Value": [200, 1000, 300, 1200]
}
df = pd.DataFrame(data)

pivot = pd.pivot_table(
    df,
    values="Value",
    index="Date",
    columns="Category",
    aggfunc="sum"
)


#print(df)

print(pivot)
```

```python
# 2. Create a Pandas DataFrame with columns 'Year', 'Quarter', and 'Revenue'. Create a pivot table to compute the mean 'Revenue' for each 'Quarter' by 'Year'.

import pandas as pd

data = {
    "Year": [2024, 2024, 2024, 2024,
             2025, 2025, 2025, 2025,
             2026, 2026, 2026, 2026],

    "Quarter": ["Q1", "Q2", "Q3", "Q4",
                "Q1", "Q2", "Q3", "Q4",
                "Q1", "Q2", "Q3", "Q4"],

    "Revenue": [100, 120, 140, 160,
                110, 130, 150, 170,
                120, 140, 160, 180]
}


#compute the mean 'Revenue' for each 'Quarter' by 'Year'.


df = pd.DataFrame(data)

pivot = pd.pivot_table(
    df,
    values="Revenue",
    index="Year",
    columns="Quarter",
    aggfunc="mean"
)

print(pivot)
```
