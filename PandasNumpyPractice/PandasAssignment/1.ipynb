```python
# 1. Create a Pandas DataFrame with 4 columns and 6 rows filled with random integers. 

# Set the index to be the first column.


import pandas as pd 

data = {
    "A":[1,2,3,4,5,6],
    "B":[7,8,9,10,11,12],
    "C":[11,12,13,14,15,18],
    "D":[16,17,18,19,20,21]

}

df = pd.DataFrame(data) ## Data Frame Created


# Set the index to be the first column.

df = df.set_index("B")

print(df)
```
