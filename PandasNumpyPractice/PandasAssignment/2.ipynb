```python
# 1. Create a Pandas DataFrame with 3 columns and 5 rows filled with random integers. Add a new column that is the product of the first two columns.

import pandas as pd 

data = {
    "A":[1,2,3,4,5],
    "B":[10,11,12,13,14],
    "C":[20,21,22,23,24]
}

df = pd.DataFrame(data)

#Add a new column that is the product of the first two columns.

df["D"] = df["A"]*df["B"]

print(df)



```

### Axis = A direction that Pandas moves 

* axis = 0 -> moves down rows-> operate on each rows 
    * 0 = down the rows, result per column

* axis = 1 -> moves across columns -> operates across columns
    * 1 = -> (across Column, result per row)

        * More Examples : 
        df.sum(axis=0)      # Column-wise sum
        df.sum(axis=1)      # Row-wise sum

        df.mean(axis=0)     # Column-wise mean
        df.mean(axis=1)     # Row-wise mean

        df.max(axis=0)      # Maximum in each column
        df.min(axis=1)      # Minimum in each row

### Deletion (axis in deletion and (drop()))

    drop + axis=0 = remove rows
    drop + axis=1 = remove columns

    sum/mean + axis=0 = column operation
    sum/mean + axis=1 = row operation


* df. drop(0,axis=0) #Delete row with index 0
* df.drop("A",axis = 1)#Delete Column A
* df.drop([0,1], axis=0) ## delete multiple rows
* df.drop(["A","B"],axis=1) ## delete multiple columns


```python
# 2. Create a Pandas DataFrame with 3 columns and 4 rows filled with random integers. Compute the row-wise and column-wise sum.

import pandas as pd 

data = {
    "A":[1,2,3,4],
    "B":[10,11,12,13],
    "C":[20,21,22,23]
}

df = pd.DataFrame(data)


## Row Wise Sum 

row_sum = df.sum(axis=1)

##Column Wise Sum

column_sum = df.sum(axis=0)


print(f"Data Frame\n{df}\nRowWiseSum\n{row_sum}\nColumnWiseSum\n{column_sum}")

```
