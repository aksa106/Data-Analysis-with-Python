# Assignment 10: Working with Text Data

1. Create a Pandas Series with 5 random text strings. Convert all the strings to uppercase.
2. Create a Pandas Series with 5 random text strings. Extract the first three characters of each string.

```python
# 1. Create a Pandas Series with 5 random text strings. Convert all the strings to uppercase.

import pandas as pd

s = pd.Series([
    "apple ",
    " banana",
    "mango",
    "orange  ",
    "grapes"

])

Upper = s.str.upper()

Lower = s.str.lower()

Strip = s.str.strip()

Title = s.str.title()

Contains = s.str.contains("a")


print(f"{Upper}\n{Lower}\n{Strip}\n{Title}\n{Contains}")
```

```python
# 2. Create a Pandas Series with 5 random text strings. Extract the first three characters of each string.

import pandas as pd

s = pd.Series([
    "apple ",
    " banana",
    "mango",
    "orange  ",
    "grapes"

])

Extract = s.str[1:3]

print(Extract)
```
