```python
import numpy as np
arr = np.array([1,2,3,4])
print(arr.shape)

## output is (4,) -> flat array with 4 elemets 
```

```python
arr = np.array([[1,2,3],[4,5,6]])
print(arr.shape)

## output is (2,3) -> 2 rows 3 columns
```

```python
arr = np.array([1,2,3,4,5,6])
## print(arr.shape)  -- (6, )

new = arr.reshape(2,3)
print(new)

## arr is flat array with 6 elements  

## reshaping it in 2 rows 3 columns 
```

###  Difference between (5,) , (1,5) , (5,1)
* (5, ) - 1D flat array with no concept of rows and columns , just a flat vector. [Dim = 1]

* (1,5) - 2D array -> 1 row 5 columns (Row vector) [Dim = 2]

* (5,1) - 2D array -> 5 rows 1 columns (column vector) [Dim = 2]



### Why is NumPy faster than Python lists?
* Contiguous memory - Numpy stores data in continous block rather than python lists which stores references (pointers)

* homogeneous Data - Arrays only stores data of same data type , unlike lists , so no type checking during operations which makes operations more faster and efficient.

* Vectorization - Numpy uses C-level loops internally m, python uses slow for-loops

* Less Overload - Python lists stores meta data for each element unlike NumPy which stores raw data values only.
