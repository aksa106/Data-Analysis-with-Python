## Numpy

Numpy is a fundamental library for scientific computing in Python.
It provides support for arrays and matrices and along with a collection of mathematical functions to operate on these data structures.

```python
import numpy as np      ## np is an alias for numpy

## Create Arrays using Numpy

#------------------------------------------------------------

# Create an 1D array
arr1 = np.array([1,2,3,4,5])

print(arr1)
print(type(arr1))
print(arr1.shape)
```

```python
## 2D array = 2 square brackets

arr2 = np.array([1,2,3,4,5])
arr2.reshape(1,5) ## 1 row and 5 columns
```

```python
## 2D array - nested list way

arr2 = np.array([[1,2,3,4,5]])
print(arr2)
print(arr2.shape)
```

```python
# 2D ARRAY
arr2 = np.array([[1,2,3,4,5],[2,3,4,5,6]])
print(arr2)
print(arr2.shape)
```

### Create Arrays with some of the In-Built Functions

#### ARANGE fucntion : -

```python

arr1 = np.arange(0,10,2)

arr2 = np.arange(0,10,2).reshape(5,1)

arr3 = np.arange(0,10,1)

print(arr1)

print(arr2)

print(arr3)
```

#### ONE function : - 


```python
np.ones ((3,4))
```

```python
## identity matrix 
np.eye(3)
```

```python
## Example
arr = np.array([[1,2,3],[4,5,6]])
print("Array: \n",arr)

print("Shape : ",arr.shape)
print(f"Number of dimensions : {arr.ndim}")
print(f"Size(number of elements): {arr.size} ")
print(f"Data type : {arr.dtype}")
print(f"Item size (in bytes) : {arr.itemsize}")
```

```python
arr1 = np.array([1,2,3,4,5])
arr2 = np.array([[[1,2,3,4]]])

shape1 = arr1.shape
shape2 = arr2.shape

print(shape1)
print(shape2 )
```
